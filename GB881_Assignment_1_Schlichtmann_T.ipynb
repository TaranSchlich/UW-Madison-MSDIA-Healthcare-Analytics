### Assigning Variables for Patient Attributes Upon Arrival

```python
# Patient demographic and clinical intake data collected at arrival
# These variables will be used throughout the notebook for calculations and decision logic

First_Name = "Albert"
Last_Name = "Cho"
Age = 45
Sex_Assigned_At_Birth = "M"
Height_In_Feet = 6.3                        # Note patient is 6.3 feet tall and not 6 feet 3 inches tall
Weight_In_Pounds = 181.39
Smoker = False
```

### Testing Data Types of Patient Attributes
### Biological Sex, Weight, Smoker Status, and Age

```python
# Verifying data types to ensure compatibility with downstream calculations
# This step helps prevent type-related errors in arithmetic and logical operations

print(type(Sex_Assigned_At_Birth))
print(type(Weight_In_Pounds))
print(type(Smoker))
print(type(Age))
```

### Calculating BMI (Body Mass Index) of Patient Using CDC BMI formula

```python
# BMI formula from CDC: BMI = (weight in pounds / height in inches²) × 703
# This formula is for clinical assessment of body composition

Height_In_Inches = Height_In_Feet * 12                         # Feet to inches conversion for a more simple BMI calculation equation
BMI = 703 * (Weight_In_Pounds / (Height_In_Inches ** 2))
BMI = round(BMI, 2)                                            # Rounding to hundreths place per assignment instruction and reassigning BMI to rounded format for later calculations
```

### Print Statement Addressing Patient by Name and Providing BMI in a Complete Sentence

```python
print(f'{First_Name} {Last_Name}, your Body Mass Index is {BMI}.')        # Called variable with format feature f-string, put variable within string by using variable name in curly brace brackets
```

### Patient Diagnosis of Bacterial Infection
### Adult Amoxicillin Dosage Calculation Rounded to Nearest Whole Mg

```python
# Adult dosage guideline: 10 mg/kg body weight
# This calculation ensures accurate dosing based on patient-specific weight

Weight_In_Kilograms = Weight_In_Pounds * 0.45                           # Converting patient weight from pounds to kilograms for correct dosage calculation. Assuming 1 pound = 0.45 kilograms
Mg_per_Kg = 10                                                          # Specifying the value of 10 as a unit of mg/kg
Adult_Dose = ((Mg_per_Kg) * (Weight_In_Kilograms))                      # Defining calculation of adult dose with patient weight in kilograms

print(f'{round(Adult_Dose)} mg')                                        # Using print and f-string to round value per assignment instruction and add units of measure for correct dosage units
```

### Total Mg of Amoxicillin for Duration of Treatment
### Patients 40-60 Years of Age Take 10 mg/kg of Patient Weight Every 8 hours for 10 Days

```python
# Calculating total number of doses based on 8-hour intervals over a 10-day treatment plan
# Ensuring adherence to prescribed regimen


Allow_Dose = (True if 40 <= Age <= 60 else False)                                 # Age verification for correct dose volume calculation
Hours_Per_Day = 24
Dosage_Frequency_In_Hours = 8
Days = 10                                                                         # Treatment interval
Total_Doses = ((Hours_Per_Day / Dosage_Frequency_In_Hours) * Days)                # Total doses over treatment period calculated by dose requirements frequency, hours per day, and over total days needed
Total_Volume = ((Total_Doses) * round(Adult_Dose))                                # Using rounded adult dose to nearest whole mg for total volume of amoxicillin


# Conditional logic to verify age eligibility for standard adult dosing
# If eligible, outputting full prescription summary for pharmacist
if Allow_Dose:
    print(f'Provide {First_Name} {Last_Name} with {round(Total_Doses)} doses of {round(Adult_Dose)} mg of amoxicillin, or {round(Total_Volume)} total mg.')
else:
    print('Age of patient is not within allowable range.')
```
